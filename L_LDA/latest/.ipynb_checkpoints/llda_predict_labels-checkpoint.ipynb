```python
import pickle
from nltk.corpus import reuters
import pandas as pd
```

```python
# loading the trained model
llda_model = pickle.load(open('llda_model.pkl', 'rb'))

# preparing test set
idlist = reuters.fileids()
idlist_test = [x for x in idlist if 'test' in x]
```

```python
# picking a test instance
id_num = 501
unseen_doc = ' '.join(reuters.words(idlist_test[id_num]))
print(unseen_doc)
```

```python
# predict labels of unseen document
topics = llda_model.inference(document=unseen_doc, iteration=200, times=10)
```

```python
df = pd.DataFrame(topics, columns =['topic', 'probability'])
df.head(10)
```

```python
# real label
reuters.categories(idlist_test[id_num])
```
