```python
import xml.etree.cElementTree as et
from bs4 import BeautifulSoup
from nltk.corpus import PlaintextCorpusReader, stopwords
from nltk.tokenize import RegexpTokenizer
from nltk.stem.wordnet import WordNetLemmatizer
from collections import defaultdict
```

```python
source = 'C:\\Users\\Terolli\\Documents\\Research Paper\\stackexchange\\stackoverflow.com-Posts\\Posts.xml'
destination_path = 'C:\\Users\\Terolli\\Documents\\Research Paper\\stackexchange\\stackoverflow.com-Posts\\'
```

```python
def extract_doc(start_date):
    prev = ['2008']

    for event, elem in et.iterparse(source):

        try:
            current = elem.attrib.get('CreationDate')[:4]
        except Exception as e:
            print(e)

        if current not in prev:
            print(current)
            prev.append(current)

        if start_date < elem.attrib.get('CreationDate'):

            # getting the body of the xml, which contains only the post with no metadata
            soup = BeautifulSoup(elem.attrib.get('Body'), 'html.parser')

            # separating natural language
            nlt_string = ''
            for para in soup.find_all('p'):
                nlt_string += para.get_text() + '\n'

            # separating code
            code_string = ''
            for code in soup.find_all('code'):
                code_string += code.get_text() + '\n'

            # writing each post as 2 separate .txt files, one for natural language and the other for code
            with open(destination_path + 'nlt\\nlt_' + elem.attrib.get('Id') + '.txt' , 'w+', encoding="utf-8") as nlt_file:
                if nlt_string != '':
                    try:
                        nlt_file.write(nlt_string)
                    except Exception as e:
                        print(e)
            with open(destination_path + 'code\\code_'+ elem.attrib.get('Id') + '.txt', 'w+', encoding="utf-8") as code_file:
                if code_string != '':
                    try:
                        code_file.write(code_string)
                    except Exception as e:
                        print(e)

            nlt_file.close()
            code_file.close()

        elem.clear()
```

```python
def preprocess(directory):
    count = 0;
    for dirpath, dirs, filenames in os.walk(directory):
        for file in filter(lambda file: file.endswith('.txt'), filenames):
            try:
                count += 1
                # convert txt file into continuous str
                document = open(os.path.join(dirpath, file),
                                encoding="utf8").read()

                # convert to lower case
                document = document.lower()
                
                # determine stop words
                stoplist = set(stopwords.words('english'))

                # tokenize document
                tk = RegexpTokenizer(r'[a-zA-Z]+')                
                
                # remove stop words
                tokens = [token for token in tk.tokenize(
                    document) if token not in stoplist]

                # remove words with length 1
                tokens = [token for token in tokens if len(token) > 1]

                # remove words with frequency == 1
                
                frequency = defaultdict(int)
                for token in tokens:
                    frequency[token] += 1
                tokens = [token for token in tokens if frequency[token] > 1]
                

                # lemmatize doc
                lemma = WordNetLemmatizer()
                tokens = [lemma.lemmatize(token) for token in tokens]
                    
                # writing each doc to file
                open('.\\corpus_stackoverflow_nlt.txt', 'a').write(' '.join(tokens) + '\n')
                
                if count%100 == 0: print(count)
                # TODO compute bigrams, trigrams

            except Exception as e:
                print(e)
```
